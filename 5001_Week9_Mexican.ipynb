```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import duckdb
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
mexi= pd.read_csv("/content/drive/MyDrive/Master_Degree/5001/Mexican.txt", sep='\t')
```

```python
mexi['item_price'] = mexi['item_price'].str.replace('$', '', regex=False).astype(float)

mx = duckdb.query("""SELECT item_name,choice_description,
                           COUNT(item_name) as item_count,
                           sum(item_price) as 'Total Dish sale'
                           FROM mexi
                           GROUP BY item_name,choice_description
                           ORDER BY item_count DESC
                  """).df()

display(mx.head())
```

```python
pd.set_option('display.max_rows', None)
```

```python
fig = px.scatter(mx,
                x='Total Dish sale',
                y='item_count',
                #hover_name='name',
                #color = 'choice_description',
                #size = 'Total Dish sale',
                hover_data=['item_name','choice_description','Total Dish sale'],
                #hover_name=  df3['Car'].astype(str)#+" : "+df3['mfr'].astype(str),
                title='Mexican')
fig.show()
```
